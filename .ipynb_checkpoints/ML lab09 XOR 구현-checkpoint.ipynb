# XOR 연산 NN으로 구현

유닛(게이트) 하나로 WX+b 를 통해 cost를 계산하고 전에했던 방식으로 학습을 시키면

accuracy가 0.5 밖에 나오질 않음

즉 여러개의 유닛을 이용해 NN을 만들어 구현하면 어떻게 될까?

```python
import tensorflow as tf
import numpy as np
```

```python
learning_rate = 0.1

x_data = np.array([[0,0],[0,1],[1,0],[1,1]], dtype=np.float32)
y_data = np.array([[0],[1],[1],[0]], dtype=np.float32)
```

```python
x = tf.placeholder(tf.float32, [None, 2])
y = tf.placeholder(tf.float32, [None, 1])

W1 = tf.Variable(tf.random_normal([2,2]), name='weight1')
b1 = tf.Variable(tf.random_normal([2]), name='bias1')
layer1 = tf.sigmoid(tf.matmul(x,W1)+b1)

W2 = tf.Variable(tf.random_normal([2,1]), name='weight2')
b2 = tf.Variable(tf.random_normal([1]), name='bias2')
hypothesis = tf.sigmoid(tf.matmul(layer1,W2)+b2)
```

```python
# cost/loss function
cost = -tf.reduce_mean(y * tf.log(hypothesis) + (1 - y) *
                       tf.log(1 - hypothesis))

train = tf.train.GradientDescentOptimizer(learning_rate=learning_rate).minimize(cost)

# Accuracy computation
# True if hypothesis>0.5 else False
predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, y), dtype=tf.float32))

# Launch graph
with tf.Session() as sess:
    # Initialize TensorFlow variables
    sess.run(tf.global_variables_initializer())

    for step in range(10001):
        sess.run(train, feed_dict={x: x_data, y: y_data})
        if step % 100 == 0:
            print(step, sess.run(cost, feed_dict={
                  x: x_data, y: y_data}), sess.run([W1, W2]))

    # Accuracy report
    h, c, a = sess.run([hypothesis, predicted, accuracy],
                       feed_dict={x: x_data, y: y_data})
    print("\nHypothesis: ", h, "\nCorrect: ", c, "\nAccuracy: ", a)
```

여기서 모델을 Wide or Deep 하게 만들면 더 성능이 좋아짐

성능이 좋아진다는 뜻은 hypothesis의 정확도가 더 올라간다는 뜻

Wide하게 하려면 각 유닛에서 나오는 y의 수를 늘려주면되고 (shape를 바꿔주는 코드 수정)

Deep하게 하려면 유닛을 몇개 더 추가하면됨 (W3,W4,b3,b4~~ 등 layer를 더 추가하는 코드 수정)
